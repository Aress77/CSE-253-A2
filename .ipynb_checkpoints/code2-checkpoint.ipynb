## Task 2 Harmonization (generate chords following a melody)

File structures: 
- `.mid` — the MIDI files, with three tracks: `MELODY` (the vocal line,
  monophonic), `BRIDGE` (a sub-melody played by a lead instrument), and
  `PIANO` (the accompaniment)
- `chord_midi.txt` — chord annotations aligned to the MIDI, one segment
  per line in the format `start_seconds end_seconds chord_label`. The labels
  follow Harte notation (`C:maj`, `A:min7`, `D:maj/5`, etc)

For this task, we condition the Melody track as input and the chord annotations
from `chord_midi.txt` as prediction targets. The `BRIDGE` and `PIANO` tracks
are not used.

### Dataset inspection

```python
def inspect_song(song_dir):
    print(f"=== Song {song_dir.name} ===\n")
    
    # Load MIDI
    midi_path = song_dir / f"{song_dir.name}.mid"
    pm = pretty_midi.PrettyMIDI(str(midi_path))
    print(f"Length: {pm.get_end_time():.1f}s, Tempo: {pm.estimate_tempo():.1f} BPM")
    print(f"Tracks:")
    for i, inst in enumerate(pm.instruments):
        print(f"  Track {i}: name='{inst.name}', {len(inst.notes)} notes")
    
    # Load chord annotations
    chord_path = song_dir / "chord_midi.txt"
    with open(chord_path) as f:
        chord_lines = f.readlines()
    print(f"\nChord annotations: {len(chord_lines)} segments")
    print("First 5 chord segments:")
    for line in chord_lines[:5]:
        parts = line.strip().split()
        if len(parts) == 3:
            start, end, label = parts
            print(f"  {float(start):6.2f}s - {float(end):6.2f}s  {label}")
```

```python
song = song_dirs[0]
pm = pretty_midi.PrettyMIDI(str(song / f"{song.name}.mid"))
melody = [inst for inst in pm.instruments if inst.name.strip().upper() == 'MELODY'][0]
melody_notes = sorted([(n.start, n.pitch) for n in melody.notes])

print(f"First melody note starts at: {melody_notes[0][0]:.2f}s")
print(f"Last melody note starts at: {melody_notes[-1][0]:.2f}s\n")

with open(song / "chord_midi.txt") as f:
    chords = [line.strip().split() for line in f]
chords = [(float(c[0]), float(c[1]), c[2]) for c in chords if len(c) == 3 and c[2] not in ('N', 'X')]

shown = 0
for start, end, label in chords:
    mel_in = [(t, p) for (t, p) in melody_notes if start <= t < end]
    if mel_in:
        print(f"Chord {label} ({start:.2f}s - {end:.2f}s):")
        for t, p in mel_in[:6]:
            print(f"  t={t:.2f}s  pitch={p} ({pretty_midi.note_number_to_name(p)})")
        print()
        shown += 1
        if shown >= 3:
            break
```

```python
def parse_song(song_dir, melody_track_name='MELODY'):

    midi_path = song_dir / f"{song_dir.name}.mid"
    chord_path = song_dir / "chord_midi.txt"
    
    if not midi_path.exists() or not chord_path.exists():
        return []
    
    pm = pretty_midi.PrettyMIDI(str(midi_path))
    
    # Find the melody track by name
    melody_inst = None
    for inst in pm.instruments:
        if inst.name.strip().upper() == melody_track_name:
            melody_inst = inst
            break
    if melody_inst is None:
        return []
    
    melody_notes = sorted([(n.start, n.pitch) for n in melody_inst.notes])
    
    rows = []
    with open(chord_path) as f:
        for i, line in enumerate(f):
            parts = line.strip().split()
            if len(parts) != 3:
                continue
            start, end, label = float(parts[0]), float(parts[1]), parts[2]
            if label in ('N', 'X'):
                continue
            
            mel_pitches = [p for (t, p) in melody_notes if start <= t < end]
            if not mel_pitches:
                continue
            
            pc = sorted(set(p % 12 for p in mel_pitches))
            rows.append({
                'song_id': song_dir.name,
                'position': i,
                'start_sec': start,
                'end_sec': end,
                'duration_sec': end - start,
                'melody_notes': ','.join(map(str, mel_pitches)),
                'melody_pitch_classes': ','.join(map(str, pc)),
                'chord_label': label,
            })
    return rows

test_rows = parse_song(song_dirs[0])
print(f"Song {song_dirs[0].name}: {len(test_rows)} (melody, chord) pairs")
print("First 3 rows:")
pd.DataFrame(test_rows[:3])
```

```python
all_rows = []
failed = 0
for i, song_dir in enumerate(song_dirs):
    try:
        all_rows.extend(parse_song(song_dir))
    except Exception as e:
        failed += 1
    if (i + 1) % 100 == 0:
        print(f"Parsed {i+1}/{len(song_dirs)}; total rows: {len(all_rows):,}; failed: {failed}")

df = pd.DataFrame(all_rows)
print(f"\nFinal: {len(df):,} rows from {df.song_id.nunique()} songs")
print(f"Failed songs: {failed}")
df.head()
```

```python
PITCH_NAMES_FLAT_TO_SHARP = {
    'Db': 'C#', 'Eb': 'D#', 'Fb': 'E', 'Gb': 'F#', 'Ab': 'G#', 'Bb': 'A#', 'Cb': 'B'
}
VALID_ROOTS = {'C','C#','D','D#','E','F','F#','G','G#','A','A#','B'}

def simplify_chord(label):
    if label in ('N', 'X') or ':' not in label:
        return None
    
    # Strip slash (bass note inversion)
    label_no_slash = label.split('/')[0]
    
    root, qual = label_no_slash.split(':', 1)
    
    # Normalize flat roots to sharps
    root = PITCH_NAMES_FLAT_TO_SHARP.get(root, root)
    if root not in VALID_ROOTS:
        return None
    
    # Bucket the quality
    qual_lower = qual.lower()
    if qual_lower.startswith('maj') or qual_lower == '':
        simple_qual = 'maj'
    elif qual_lower.startswith('min'):
        simple_qual = 'min'
    elif qual_lower in ('7', 'dom7', '9', '11', '13'):
        simple_qual = 'dom7'
    elif qual_lower.startswith('dim') or qual_lower == 'hdim7':
        simple_qual = 'dim'
    elif qual_lower.startswith('aug'):
        simple_qual = 'aug'
    elif qual_lower.startswith('sus'):
        simple_qual = 'sus'
    else:
        simple_qual = 'other'
    
    return f'{root}:{simple_qual}'

# Apply to all rows
df['chord_simple'] = df['chord_label'].apply(simplify_chord)
df['chord_root'] = df['chord_simple'].str.split(':').str[0]
df['chord_quality'] = df['chord_simple'].str.split(':').str[1]

# Drop any rows that couldn't be simplified
before = len(df)
df = df.dropna(subset=['chord_simple']).reset_index(drop=True)
print(f"Dropped {before - len(df)} rows with un-parseable chords")

print(f"\nOriginal distinct chord labels: {df['chord_label'].nunique()}")
print(f"Simplified distinct chord labels: {df['chord_simple'].nunique()}")
print(f"\nTop 15 simplified chords:")
print(df['chord_simple'].value_counts().head(15))
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 9))

# 1. Top 20 chord labels
top20 = df['chord_simple'].value_counts().head(20)
top20.plot(kind='bar', ax=axes[0, 0], color='steelblue')
axes[0, 0].set_title('Top 20 chord labels in POP909')
axes[0, 0].set_ylabel('count')
axes[0, 0].tick_params(axis='x', rotation=45)

# 2. Chord quality breakdown
qual_counts = df['chord_quality'].value_counts()
qual_counts.plot(kind='bar', ax=axes[0, 1], color='coral')
axes[0, 1].set_title('Chord quality distribution')
axes[0, 1].set_ylabel('count')
axes[0, 1].tick_params(axis='x', rotation=0)
for i, v in enumerate(qual_counts.values):
    axes[0, 1].text(i, v, f'{100*v/len(df):.1f}%', ha='center', va='bottom', fontsize=9)

# 3. Melody notes per chord segment
df['n_melody_notes'] = df['melody_notes'].str.count(',') + 1
n_clipped = df['n_melody_notes'].clip(upper=20)
axes[1, 0].hist(n_clipped, bins=range(1, 22), color='seagreen', edgecolor='white')
axes[1, 0].set_title('Melody notes per chord segment (clipped at 20)')
axes[1, 0].set_xlabel('# notes')
axes[1, 0].set_ylabel('# segments')
axes[1, 0].axvline(df['n_melody_notes'].median(), color='red', linestyle='--',
                    label=f"median = {df['n_melody_notes'].median():.0f}")
axes[1, 0].legend()

# 4. Chord duration distribution
dur_clipped = df['duration_sec'].clip(upper=6)
axes[1, 1].hist(dur_clipped, bins=30, color='mediumpurple', edgecolor='white')
axes[1, 1].set_title('Chord segment duration (clipped at 6s)')
axes[1, 1].set_xlabel('duration (seconds)')
axes[1, 1].set_ylabel('# segments')
axes[1, 1].axvline(df['duration_sec'].median(), color='red', linestyle='--',
                    label=f"median = {df['duration_sec'].median():.2f}s")
axes[1, 1].legend()

plt.tight_layout()
plt.savefig(OUT_DIR / 'eda_overview.png', dpi=120, bbox_inches='tight')
plt.show()

print(f"\nSummary stats:")
print(f"  Total pairs: {len(df):,}")
print(f"  Total songs: {df.song_id.nunique()}")
print(f"  Vocab size: {df.chord_simple.nunique()}")
print(f"  Median melody notes per segment: {df['n_melody_notes'].median():.0f}")
print(f"  Median chord duration: {df['duration_sec'].median():.2f}s")
print(f"  Most common chord: {df['chord_simple'].mode()[0]} ({df['chord_simple'].value_counts().iloc[0]:,} occurrences = {100*df['chord_simple'].value_counts().iloc[0]/len(df):.1f}%)")
```

```python
songs = df['song_id'].unique()
rng = np.random.default_rng(42)
rng.shuffle(songs)

n = len(songs)
train_songs = set(songs[:int(0.8 * n)])
val_songs   = set(songs[int(0.8 * n):int(0.9 * n)])
test_songs  = set(songs[int(0.9 * n):])

df['split'] = df['song_id'].map(
    lambda s: 'train' if s in train_songs else 'val' if s in val_songs else 'test'
)

print(f"Songs:  train={len(train_songs)}, val={len(val_songs)}, test={len(test_songs)}")
print(f"Rows:\n{df['split'].value_counts()}")
print(f"\nFraction per split:\n{df['split'].value_counts(normalize=True).round(3)}")
```

### Baseline

```python
train = df[df['split'] == 'train']
test  = df[df['split'] == 'test']
vocab = sorted(train['chord_simple'].unique())

print(f"Train rows: {len(train):,}")
print(f"Test rows: {len(test):,}")
print(f"Chord vocab size: {len(vocab)}")

# 1. Random
def baseline_random(test_df, seed=0):
    rng = random.Random(seed)
    return [rng.choice(vocab) for _ in range(len(test_df))]

# 2. Most common chord
most_common_chord = train['chord_simple'].mode()[0]
def baseline_most_common(test_df):
    return [most_common_chord] * len(test_df)

# 3: Melody pitch class
pc_to_chord = (
    train.groupby('melody_pitch_classes')['chord_simple']
         .agg(lambda s: s.mode().iloc[0])
         .to_dict()
)
fallback_chord = most_common_chord

def baseline_pc_lookup(test_df):
    return [pc_to_chord.get(pc, fallback_chord) for pc in test_df['melody_pitch_classes']]

print(f"\nMost common chord in training: {most_common_chord}")
print(f"Distinct melody pitch-class sets in training: {len(pc_to_chord):,}")
print(f"\nExample predictions for first 5 test rows:")
sample = test.head(5)[['melody_pitch_classes', 'chord_simple']]
sample['random'] = baseline_random(sample)
sample['most_common'] = baseline_most_common(sample)
sample['pc_lookup'] = baseline_pc_lookup(sample)
sample
```

### Eval?

```python
from collections import Counter

def confusion_top_n(predictions, truth, n=10):
    """For the n most common true chords, show what each baseline predicted."""
    top_chords = [c for c, _ in Counter(truth).most_common(n)]
    rows = []
    for true_chord in top_chords:
        mask = [t == true_chord for t in truth]
        preds_for_this = [p for p, m in zip(predictions, mask) if m]
        pred_dist = Counter(preds_for_this).most_common(3)
        rows.append({
            'true_chord': true_chord,
            'count': sum(mask),
            'top_predictions': ', '.join([f'{c}({n})' for c, n in pred_dist])
        })
    return pd.DataFrame(rows)

print("PC-lookup baseline: what did it predict for the 10 most common true chords?\n")
truth_list = test['chord_simple'].tolist()
preds_pc = baseline_pc_lookup(test)
print(confusion_top_n(preds_pc, truth_list, n=10).to_string(index=False))
```
